```python
# Check missing values

df.isnull().sum()
```

```python
# Remove missing values

df = df.dropna()

print("Missing values removed successfully.")
```

```python
# Check duplicate rows

print("Duplicate Rows :", df.duplicated().sum())
```

```python
# Remove duplicate rows

df = df.drop_duplicates()

print("Duplicate rows removed successfully.")
```

```python
# Convert discounted_price into numeric

df["discounted_price"] = (
    df["discounted_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df["actual_price"] = (
    df["actual_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df["discount_percentage"] = (
    df["discount_percentage"]
    .str.replace("%", "", regex=False)
    .astype(float)
)
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

```python
df["rating_count"] = (
    df["rating_count"]
    .str.replace(",", "", regex=False)
)

df["rating_count"] = pd.to_numeric(
    df["rating_count"],
    errors="coerce"
)
```

```python
df.info()
```

```python
# Remove rows where rating is missing
df = df.dropna(subset=["rating"])

print("Updated Shape:", df.shape)
```

```python
print("Total Products :", len(df))
print("Average Rating :", round(df["rating"].mean(), 2))
print("Highest Rating :", df["rating"].max())
print("Lowest Rating :", df["rating"].min())
```

```python
print("Average Discounted Price :", round(df["discounted_price"].mean(), 2))
print("Average Actual Price :", round(df["actual_price"].mean(), 2))
print("Highest Product Price :", df["actual_price"].max())
print("Lowest Product Price :", df["actual_price"].min())
```

```python
plt.figure(figsize=(8,5))

df["main_category"].value_counts().plot(kind="bar")

plt.title("Number of Products by Category")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.xticks(rotation=30)

plt.tight_layout()

plt.savefig("output/charts/category_count.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["rating"], bins=10)

plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.tight_layout()

plt.savefig("output/charts/rating_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["discount_percentage"], bins=20)

plt.title("Discount Percentage Distribution")
plt.xlabel("Discount %")
plt.ylabel("Products")

plt.tight_layout()

plt.savefig("output/charts/discount_distribution.png")

plt.show()
```

```python
top10 = df.nlargest(10, "actual_price")

plt.figure(figsize=(12,6))

plt.bar(top10["product_name"], top10["actual_price"])

plt.title("Top 10 Most Expensive Products")
plt.xlabel("Product")
plt.ylabel("Price")

plt.xticks(rotation=90)

plt.tight_layout()

plt.savefig("output/charts/top_expensive_products.png")

plt.show()
```

```python
rating_category = (
    df.groupby("main_category")["rating"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

plt.bar(rating_category.index, rating_category.values)

plt.title("Average Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")

plt.xticks(rotation=30)

plt.tight_layout()

plt.savefig("output/charts/average_rating_category.png")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["actual_price"],
    df["discounted_price"],
    alpha=0.5
)

plt.title("Actual Price vs Discounted Price")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")

plt.tight_layout()

plt.savefig("output/charts/price_scatter.png")

plt.show()
```

```python
top_review = df.nlargest(10, "rating_count")

plt.figure(figsize=(12,6))

plt.bar(top_review["product_name"], top_review["rating_count"])

plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Product")
plt.ylabel("Reviews")

plt.xticks(rotation=90)

plt.tight_layout()

plt.savefig("output/charts/top_reviewed_products.png")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("output/charts/heatmap.png")

plt.show()
```
